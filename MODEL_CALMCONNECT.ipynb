```python
for i in ["Code"]:
    df[i].fillna(df[i].mode(),inplace=True)
```

```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include="number").columns:
    df[i]=impute.fit_transform((df[[i]]))
```

```python
#Outliers treatment
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
df.columns
```

```python
wisker(df['Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)'])
```

```python
for i in ['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)','Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
for i in ['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)','Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)']:
    sns.boxplot(df[i])
    plt.show()
```

```python
# Define thresholds for binary classification for anxiety and depressive disorders

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Initialize the scaler
scaler = StandardScaler()
anxiety_threshold = 4.8
depression_threshold = 5.0

# Create target columns for each condition
df['anxiety_target'] = (df['Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)'] > anxiety_threshold).astype(int)
df['depression_target'] = (df['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)'] > depression_threshold).astype(int)

# We will use the same features (other disorder prevalences) for both models
# Define the features (columns used as input for prediction)
features = [
    'Prevalence - Bipolar disorder - Sex: Both - Age: Age-standardized (Percent)',
    'Prevalence - Eating disorders - Sex: Both - Age: Age-standardized (Percent)',
    'Prevalence - Drug use disorders - Sex: Both - Age: Age-standardized (Percent)',
    'Prevalence - Alcohol use disorders - Sex: Both - Age: Age-standardized (Percent)'
]

# Now you can proceed with selecting the features from the dataframe
X = df[features]

# Split the data for anxiety disorder prediction
y_anxiety = df['anxiety_target']
X_train_anxiety, X_test_anxiety, y_train_anxiety, y_test_anxiety = train_test_split(X, y_anxiety, test_size=0.2, random_state=42)

# Split the data for depressive disorder prediction
y_depression = df['depression_target']
X_train_depression, X_test_depression, y_train_depression, y_test_depression = train_test_split(X, y_depression, test_size=0.2, random_state=42)

# Train models for both conditions
# Scaling the features
X_train_anxiety_scaled = scaler.fit_transform(X_train_anxiety)
X_test_anxiety_scaled = scaler.transform(X_test_anxiety)

# Scale the features for depression prediction
X_train_depression_scaled = scaler.fit_transform(X_train_depression)
X_test_depression_scaled = scaler.transform(X_test_depression)
# Train the logistic regression models
model_anxiety = LogisticRegression()
model_anxiety.fit(X_train_anxiety_scaled, y_train_anxiety)

model_depression = LogisticRegression()
model_depression.fit(X_train_depression_scaled, y_train_depression)

# Make predictions
y_pred_anxiety = model_anxiety.predict(X_test_anxiety_scaled)
y_pred_depression = model_depression.predict(X_test_depression_scaled)

# Calculate accuracies for both models
accuracy_anxiety = accuracy_score(y_test_anxiety, y_pred_anxiety)
accuracy_depression = accuracy_score(y_test_depression, y_pred_depression)

accuracy_anxiety, accuracy_depression
```
